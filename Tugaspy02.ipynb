```
nim = ["10121001", "10121002", "10121003"]
nama = ["Asep", "Budi", "Cecep"]

nilai = [
    [50, 70, 40, 80],
    [78, 78, 80, 65],
    [57, 88, 67, 69]
]

rata_mahasiswa = []
for n in nilai:
    rata = sum(n) / len(n)
    rata_mahasiswa.append(rata)

maks = max(rata_mahasiswa)
index_mhs = rata_mahasiswa.index(maks)

rata_mk = []
for j in range(len(nilai[0])):
    total = 0
    for i in range(len(nilai)):
        total += nilai[i][j]
    rata = total / len(nilai)
    rata_mk.append(rata)

min_mk = min(rata_mk)
index_mk = rata_mk.index(min_mk)

print("Mahasiswa Terpintar :")
print("NIM   :", nim[index_mhs])
print("Nama  :", nama[index_mhs])
print("Nilai :", round(maks, 2))
print("==============================")
print("Mata Kuliah Nilai Terkecil :")
print("MK", index_mk + 1)
print("Nilai :", round(min_mk, 2))
```

```
def input_matriks(b, k):
    return [[int(input(f"[{i}][{j}]: ")) for j in range(k)]
            for i in range(b)]
def tampil(m):
    for r in m: print(r)
def tambah(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))]
            for i in range(len(A))]
def kurang(A, B):
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))]
            for i in range(len(A))]
def kali(A, B):
    return [[sum(A[i][k] * B[k][j] for k in range(len(B)))
            for j in range(len(B[0]))] for i in range(len(A))]

while True:
    print("Pilih Menu yang tersedia :")
    print("1.Tambah")
    print("2.Kurang")
    print("3.Kali")
    print("0.Exit")
    p = input("Pilih: ")

    if p in ["1", "2"]:
        b = int(input("Baris[]: "))
        k = int(input("Kolom[]: "))
        print("Matriks A"); A = input_matriks(b, k)
        print("Matriks B"); B = input_matriks(b, k)
        hasil = tambah(A, B) if p == "1" else kurang(A, B)
        print("Hasil:"); tampil(hasil)

    elif p == "3":
        bA = int(input("Baris A: "))
        kA = int(input("Kolom A: "))
        bB = int(input("Baris B: "))
        kB = int(input("Kolom B: "))
        if kA != bB:
            print("Silakan Ganti Angka!")
        else:
            print("Matriks A"); A = input_matriks(bA, kA)
            print("Matriks B"); B = input_matriks(bB, kB)
            print("Hasil:"); tampil(kali(A, B))
    elif p == "0":
        print("BYEE ANAK MANIS")
        break
```
